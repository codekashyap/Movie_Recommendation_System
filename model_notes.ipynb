```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credit = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies_merged = pd.merge(movies, credit, on=['title'])
```

```python
movies_merged.columns
```

```python
df1= movies_merged.loc[: , ('id', 'title','genres', 'keywords', 'overview', 'cast', 'crew')]
df1.sample(2)
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace=True)
```

```python
print(df1.iloc[1539].genres)
type(df1.iloc[1539].genres)
```

```python
# to convert string of list into list use function -> ast.literal_eval(str)
import ast
def change(l):
    L = []
    for i in ast.literal_eval(l):
        L.append(str(i['name']))
    return L
```

```python
df1.loc[:,'keywords'] = df1.loc[:,'keywords'].apply(change)
df1.loc[: ,'genres'] = df1.loc[:, 'genres'].apply(change)
```

```python
# to get top 3 cast 
def change3(l):
    L = []
    for i in ast.literal_eval(l)[0:3]:
        L.append(str(i['name']))
    return L
df1.loc[:,'cast'] = df1.loc[:,'cast'].apply(change3)
```

```python
# to get director name
def get_dir(l):
    L = []
    for i in ast.literal_eval(l):
        if i['job'] == "Director":
            L.append(i['name'])
    return L
df1.loc[:,'crew'] = df1.loc[:,'crew'].apply(get_dir)
```

```python
df1
```

```python
#remove space between words in order to create better tags
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df1.loc[:,'crew'] = df1.loc[:,'crew'].apply(remove_space)
df1.loc[:,'genres'] = df1.loc[:,'genres'].apply(remove_space)
df1.loc[:,'keywords'] = df1.loc[:,'keywords'].apply(remove_space)
df1.loc[:,'cast'] = df1.loc[:,'cast'].apply(remove_space)
```

```python
df1
```

```python
df1.loc[:,'overview'] = df1.loc[:,'overview'].apply(lambda x : x.split())
```

```python
df1
```

```python
df1['tags'] = df1['overview'] + df1['genres'] + df1['keywords'] + df1['cast'] + df1['crew']
df1['tags'] = df1['tags'].apply(lambda x: " ".join(x))
```

```python
df1['tags'] = df1['tags'].apply(lambda x:x.lower())
```

```python
df1['tags'][0]
```

```python
df2 = df1.loc[: , ["id", "title", "tags"]]
df2
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
#example
stem('loves loved')
```

```python
df2.loc[:,'tags'] = df2.loc[:,'tags'].apply(stem)
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features= 5000 , stop_words="english")
```

```python
vector = cv.fit_transform(df2['tags'])
```

```python
type(vector)
```

```python
vector = vector.toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
#similarity btw first and 1 to 4806 movies
similarity[0]
```

```python
def recommend(movie):
    index = df2[df2['title'] == movie].index[0]
    # using enumarate to create tuple of similarity with index
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    for i in distances[1:6]:
        print(df2.iloc[i[0]].title)
```

```python
try:
    recommend(input("Enter the movie name: "))
except IndexError:
    print("Movie is not in our database. Sorry!")
```
